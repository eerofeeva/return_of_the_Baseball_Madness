```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
attendance_Det=[38067,36015,33655]
x=[0,1,2]
y_limit = 39000
y_axis = np.arange(32000, y_limit, 1)
plt.ylim(32000, y_limit)
years=['2013','2014','2015']
plt.xticks(x, years, rotation='vertical')
plt.margins(0.2)
plt.subplots_adjust(bottom=.15)
plt.plot(attendance_Det)
plt.ylabel("DET Avg Attendance 2013-2015")
plt.show()
```
